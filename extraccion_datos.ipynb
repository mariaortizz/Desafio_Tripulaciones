```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
import numpy as np
```

```python
url = "https://www.meraparivar.org/financial-accounts/"
response = requests.get(url)
```

```python
html = response.content
```

```python
soup = bs(html, "html.parser")
```

```python
print(soup)
```

```python
table = soup.find_all('table', class_ = "tablepress tablepress-id-1")
```

```python
table
```

```python
data = {}
for i in range(1,6):
    columna = table[0].find_all('th', class_ = f"column-{i}")[0].text
    if columna != '':
        data[columna] = []
    else:
        data['Country'] = []
```

```python
data
```

```python
for index, i in enumerate(data.keys()):
    columna_elementos = table[0].find_all('td', class_ = f"column-{index+1}")
    for a in columna_elementos:
        data[i].append(a.text)
```

```python
for i in data:
    print(i)
    print(len(data[i]))
```
